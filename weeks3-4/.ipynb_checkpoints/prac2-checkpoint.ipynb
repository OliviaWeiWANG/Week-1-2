1\.Write a program to solve a classic ancient Chinese puzzle:  
 We count 35 heads and 94 legs among the chickens and rabbits in a farm. How many rabbits and how many chickens do we have? 

````python
##Hint:Use for loop to iterate all possible
## write your answer here!
#when legs = 94 while heads = 35, there are 23 chicken and 12 rabbits
legs = int(input("input the number of legs: "))
heads = int(input("input the number of heads: "))
for chicken in range(0, heads):
    for rabbit in range(0, heads):
        if (2 * chicken + 4 * rabbit) == legs and (chicken + rabbit) == heads:
            print("There are {} chickens， {} rabbits".format(chicken, rabbit))
````

2\. _Social networking_ Please look at the graph below of social relationships. You have a mango farm, you need to find a mango buyer and sell him your mango that you have grown (just a small farm). To this end, our principle is to use our social network of friends separated by one degree, and then second degree friends (friends of friends), third degree friends (friends of friends of friends).

Use the depth first algorithm to search your network until you find a mango seller. 

We make an assumption that if the person's name ends with m, then it means the person must be the mango seller!


<img src="img/p3.png" width="500" height="400" >

````python
## write your answer here!
class Graph(object):

    def __init__(self):
        self.nodeNeighbor = {}
        self.visited = {}

    def nodes(self):
        return self.nodeNeighbor.keys()

    def add_node(self, node):
        if node not in self.nodes():
            self.nodeNeighbor[node] = []

    def initial(self, nodelist):
        for node in nodelist:
            self.add_node(node)

    def add_edge(self, edge):
        u, v = edge
        if (u not in self.nodeNeighbor[v]) and (v not in self.nodeNeighbor[u]):
            self.nodeNeighbor[u].append(v)
            if u != v:
                self.nodeNeighbor[v].append(u)

    def depth_first_search(self, root):
        namelist = []

        def dfs(person):
            self.visited[person] = True
            namelist.append(person)
            for n in self.nodeNeighbor[person]:
                if n[-1] == 'm':
                    namelist.append(n)
                    return
                if n not in self.visited:
                    dfs(n)
        if root:
            dfs(root)
        for node in self.visited:
            if node not in self.visited:
                dfs(node)
        return namelist


nodes = ['you', 'anuj', 'bob', 'peggy', 'alice', 'claire', 'thom', 'jonny']
g = Graph()
g.initial(nodes)
g.add_edge(('bob','anuj'))
g.add_edge(('bob','peggy'))
g.add_edge(('bob','you'))
g.add_edge(('peggy','alice'))
g.add_edge(('you','alice'))
g.add_edge(('you','claire'))
g.add_edge(('claire','thom'))
g.add_edge(('claire','jonny'))
seller = ''
order = g.depth_first_search('you')
print("The search order is ", order)
print("The mango seller is " + order[-1])
````

3\. Depth First traverse：Use depth first traversal to traverse all the DLL files on the C disk and output the file names.

Update your code to search for a particular filename.

What protocol did you use to select the next node to expand on the fringe?

````python
## write your answer here!
import os


def dfs_dir(path):
    stack = [path]
    visited = []
    while len(stack) > 0:
        try:
            temp = stack.pop(len(stack) - 1)
            if os.path.isdir(temp):
                visited.append(temp)
                for d in os.listdir(temp):
                    stack.append(os.path.join(temp, d))
            elif os.path.isfile(temp):
                visited.append(temp)
                if get_suffix(temp) == '.dll':
                    print(os.path.basename(os.path.realpath(temp)))
        except PermissionError:
            pass
    return visited


def get_suffix(path):
    return os.path.splitext(path)[1]


print("All the DLL files on the C disk:")
dfs_dir('C:/')
````

4\.
Breadth First traverse：Use breadth traversal to traverse all the DLL files on the C disk and output the filename.

Update your code to search for a particular filename.

````python
## write your answer here!
import os


def bfs_dir(path):
    queue = [path]
    visited = []
    while len(queue) > 0:
        try:
            temp = queue.pop(0)
            if os.path.isdir(temp):
                visited.append(temp)
                for d in os.listdir(temp):
                    queue.append(os.path.join(temp, d))
            elif os.path.isfile(temp):
                visited.append(temp)
                if get_suffix(temp) == '.dll':
                    print(os.path.basename(os.path.realpath(temp)))
        except PermissionError:
            pass
    return visited


def get_suffix(path):
    return os.path.splitext(path)[1]


print("All the DLL files on the C disk:")
bfs_dir('C:/')
````

5\. Uniform cost search

Consider a state space where the start state is 2 and each state k has three successors: numbers 2k, 2k+1, 2k+2. The cost from state k to each respective child is k, ground(k/2), k+2.

can you implement a uniform-cost-search algorithm with python. The goal state is number 85.

````python
goal = 85
state = [-1] * (goal + 1)
state[2] = 0
for k in range(2, int(goal/2) + 1):
    if state[k] < 0:
        continue
    for pos, cost in [
        (k*2, k),
        (k*2+1, k//2),
        (k*2+2, k+2)]:
        if pos > goal:
            continue
        if state[pos] == -1 or state[pos] > state[k] + cost:
            state[pos] = state[k] + cost
print(state[goal])
````
